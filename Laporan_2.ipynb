```python
#202331161_Agustne
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Membaca Gambar

```python
#202331161_Agustne
img = cv2.imread('mcd.webp')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Citra Asli")
plt.axis("off")
plt.show()
```

# Menampilkan Citra

```python
#202331161_Agustne
print("Dimensi Gambar :", img.shape)
print("Tipe Data :", img.dtype)
```

# Konversi RGB ke Grayscale

```python
#202331161_Agustne
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis("off")
plt.show()
```

# Operasi Ketetanggaan 4

```python
#202331161_Agustne
x = 100
y = 100

print("Pixel Pusat :", gray[x,y])

print("Atas :", gray[x-1,y])
print("Bawah :", gray[x+1,y])
print("Kiri :", gray[x,y-1])
print("Kanan :", gray[x,y+1])
```

# Operasi Ketetanggaan 8

```python
#202331161_Agustne
tetangga8 = gray[x-1:x+2, y-1:y+2]

print(tetangga8)
```

# Average Filter (Konvolusi)

```python
#202331161_Agustne
average = cv2.blur(gray,(3,3))

plt.imshow(average,cmap='gray')
plt.title("Average Filter")
plt.axis("off")
plt.show()
```

# Median Filter

```python
#202331161_Agustne
median = cv2.medianBlur(gray,3)

plt.imshow(median,cmap='gray')
plt.title("Median Filter")
plt.axis("off")
plt.show()
```

# Menambahkan Gaussian Noise

```python
#202331161_Agustne
row,col = gray.shape

gaussian = np.random.normal(0,25,(row,col))

gaussian_img = gray + gaussian

gaussian_img = np.clip(gaussian_img,0,255).astype(np.uint8)

plt.imshow(gaussian_img,cmap='gray')
plt.title("Gaussian Noise")
plt.axis("off")
plt.show()
```

# Menghilangkan Gaussian Noise dengan Average Filter

```python
#202331161_Agustne
hasil_average = cv2.blur(gaussian_img,(3,3))

plt.imshow(hasil_average,cmap='gray')
plt.title("Gaussian Noise + Average Filter")
plt.axis("off")
plt.show()
```

# Edge Detection (Konvolusi)

```python
#202331161_Agustne
sobelx = cv2.Sobel(gray,
                   cv2.CV_64F,
                   1,
                   0,
                   ksize=3)

plt.imshow(sobelx,cmap='gray')
plt.title("Deteksi Tepi Sobel")
plt.axis("off")
plt.show()
```

# Menampilkan Semua Hasil

```python
#202331161_Agustne
plt.figure(figsize=(12,8))

plt.subplot(221)
plt.imshow(gray,cmap='gray')
plt.title('Asli')

plt.subplot(222)
plt.imshow(average,cmap='gray')
plt.title('Average')

plt.subplot(223)
plt.imshow(median,cmap='gray')
plt.title('Median')

plt.subplot(224)
plt.imshow(sobelx,cmap='gray')
plt.title('Sobel')

plt.show()
```
